```python
#Abre una imagen, muestra shape, min, max y promedio de cada canal
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("imagen.jpg")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
print("Shape:", img.shape)
```

```python
# Verificar si la imagen tiene canales (3 o 4) o es escala de grises (2D)
if img.ndim == 2:
    # Imagen en escala de grises
    print(f"Min: {img.min()}")
    print(f"Max: {img.max()}")
    print(f"Promedio: {img.mean():.2f}")
else:
    # Imagen con canales
    canales = ['Rojo', 'Verde', 'Azul']
    if img.shape[2] == 4:
        canales.append('Alfa')

    for i, canal in enumerate(canales):
        min_val = img[:, :, i].min()
        max_val = img[:, :, i].max()
        mean_val = img[:, :, i].mean()
        print(f"{canal} - Min: {min_val}, Max: {max_val}, Promedio: {mean_val:.2f}")
```
