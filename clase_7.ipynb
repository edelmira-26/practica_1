```python
# ¿Qué proveedores tienen más quejas registradas?
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Quejas.xlsx")
```

```python
df.head()
```

```python
# Contar quejas por proveedor
```

```python
Quejas_proveedor = df["PROVEEDOR"].value_counts().head(10)
print(Quejas_proveedor)
```

```python
#grafico
```

```python
plt.figure(figsize=(10,6))
Quejas_proveedor.plot(kind='bar', color='skyblue')
plt.title('Top 10 Proveedores con Más Quejas')
plt.xlabel('Proveedor')
plt.ylabel('Número de Quejas')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Proveedores con Más Quejas Registradas
#El análisis muestra que CFE Suministrador de Servicios Básicos concentra el mayor número de quejas,seguido por Walmart México y Aerovias. Esto indica que los sectores de servicios públicos, comercio minorista y transporte aéreo generan la mayor cantidad de inconformidades por parte de los consumidores
```
